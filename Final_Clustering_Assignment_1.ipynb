### Hopkins Statistics:

```python
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
import numpy as np
from math import isnan

def hopkins(X):
    d = X.shape[1]
    #d = len(vars) # columns
    n = len(X) # rows
    m = int(0.1 * n) 
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)
 
    rand_X = sample(range(0, n, 1), m)
 
    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
 
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd, wjd)
        H = 0
 
    return H
```

# K Means Clustering

### Finding the Optimal Number of Clusters

#### SSD

```python
# elbow-curve/SSD
ssd = []
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(df_country_scaled)
    
    ssd.append(kmeans.inertia_)
    
# plot the SSDs for each n_clusters
# ssd
plt.plot(ssd)
```

```python
# silhouette analysis
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]

for num_clusters in range_n_clusters:
    
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(df_country_scaled)
    
    cluster_labels = kmeans.labels_
    
    # silhouette score
    silhouette_avg = silhouette_score(df_country_scaled, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
    
```

```python
df1 = pd.DataFrame(df_country_scaled, columns = country_data.columns[1:])
df1.head()
```

```python
from sklearn.metrics import silhouette_score
ss = []
for k in range(2, 11):
    kmean = KMeans(n_clusters = k).fit(df1)
    ss.append([k, silhouette_score(df1, kmean.labels_)])
temp = pd.DataFrame(ss)    
plt.plot(temp[0], temp[1])
```

```python
# final model with k=3
kmeans = KMeans(n_clusters=3, max_iter=50,random_state=0)
kmeans.fit(df_country_scaled)
```

```python
kmeans.labels_
```

## Hierarchical Clustering

```python
# Dropping the old cluster label and saving in new datset
hier =country_data.drop('cluster_id',1)
hier.head()
```

```python
hier1 =hier.drop('country',1)
```

```python
hier1.head()
```

```python
# single linkage
mergings = linkage(hier1, method="single", metric='euclidean')
dendrogram(mergings)
plt.show()
```

As it is not showing cluster properly I will go for complete linkage

```python
plt.figure(figsize=(12,8))            
sns.set_style('white')                  
plt.title('Hierarchical Clustering Dendrogram',fontsize=20)
plt.xlabel('Country',fontsize=20)
plt.ylabel('pc value',fontsize=20)
mergings = linkage(df_country_scaled, method="complete", metric='euclidean')
dendrogram(mergings,labels=hier1.index)
plt.show()
```

```python
# 3 clusters
hier_label=cut_tree(mergings,n_clusters=3).reshape(-1, )
hier_label
```

```python
hier['hier_label'] = hier_label
```
